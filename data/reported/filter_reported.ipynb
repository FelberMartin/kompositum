```python
import pandas as pd
```

```python
# Read the data from the all_reports.csv file 
df = pd.read_csv('all_reported.csv', names=["compound", "modifier", "head"], keep_default_na=False)
df
```

```python
main_df = pd.read_csv('../../assets/filtered_compounds.csv', keep_default_na=False)
main_df
```

```python
# Print the compounds that are already in the filtered_compounds.csv file
df_in_both = df.merge(main_df, on=['compound'])
df_in_both
```

```python
# Filter the compounds from df_in_both where modifier_x is not equal to modifier_y or head_x is not equal to head_y
df_different = df_in_both[(df_in_both['modifier_x'] != df_in_both['modifier_y']) | (df_in_both['head_x'] != df_in_both['head_y'])]
df_different
```

```python
# Combine df_different with the rows from df that are not in df_main
df_different = df_different[['compound', 'modifier_x', 'head_x']]
df_different.columns = ['compound', 'modifier', 'head']
df_new = df_different.append(df[~df['compound'].isin(df_in_both['compound'])])
df_new
```

```python
# Sort the rows by compound
df_sorted = df_new.sort_values(by=['compound'])
df_sorted
```

```python
# Save the new dataframe to a csv file
df_sorted.to_csv('../../assets/reported_compounds.csv', index=False)
```
